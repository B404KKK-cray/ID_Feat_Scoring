```python
import pandas as pd
import numpy as np
from optbinning import OptimalBinning
```

```python
datas = pd.read_csv(r'C:\Users\kevin.puika\OneDrive - Bank Jasa Jakarta\Desktop\Saku Kredit Query Jobs\02. Feat Eng\Full_data_mdl.csv')

datas['dpd_os_14_acct'] = datas['dpd_os_14_acct'].fillna(0).astype(np.int64)
datas['dpd_os_30_acct'] = datas['dpd_os_30_acct'].fillna(0).astype(np.int64)
datas['disb_amt'] = datas['disb_amt'].fillna(0).astype(np.int64)
```

```python
datas.head()
```

```python
variable = "age"
x = datas[variable].values
y = datas.dpd_os_30_acct
```

```python
optb = OptimalBinning(name=variable, dtype="numerical", solver="cp")

## dtype: categorical, numerical
```

```python
optb.fit(x, y)
```

```python
optb.status
```

```python
optb.splits
```

```python
binning_table = optb.binning_table
type(binning_table)
binning_table.build()
```
